> # **Como funciona o Google Apps Script?**

> ## O que é Google Apps Script?

O **Google Apps Script** é uma plataforma de desenvolvimento de aplicativos que torna mais fácil e rápido criar aplicativos que se integram ao Google Workspace.

Você escreve código em JavaScript moderno e tem acesso a bibliotecas integradas para os aplicativos favoritos do Google Workspace, como Gmail, Agenda, Drive e muito mais. Não há nada para instalar - usa-se um editor de código direto no seu navegador e seus scripts são executados nos **servidores do Google**.

> ## O que é Javascript?

Esse tópico é para os novos programadores que terão o primeiro contato com programação utilizando o Google Apps Script. Dissemos que os códigos escritos na plataforma de desenvolvimento da Google são feitos com Javascript, mas **o que é Javascript?**

JavaScript é uma linguagem de programação de alto nível criada, a princípio, para ser executada em navegadores e manipular comportamentos de páginas web.


Segundo a Mozilla Foundation, atual nome da antiga Netscape Communications Corporations, empresa responsável pela criação do JS, "JavaScript é uma linguagem de programação, leve, interpretada, orientada a objetos, baseada em protótipos e em first-class functions (funções de primeira classe), mais conhecida como a linguagem de script da Internet."

> ## Comentários

Vamos começar falando sobre comentários. Para começar, você deve escrever o código dentro da função ```myFunction()```. 

image.png

Um comentário é uma parte de código que não é interpretada pelo navegador. Pode ser utilizada, também, para colocar textos que ajudam a identificar o que está sendo feito naquele trecho de código. Pensando num trabalho em equipe, esta é uma excelente prática.

Existem dois tipos de comentários:

> ### Comentário com duas Barras ```//```

Serve para comentar uma linha de código. 

```
function myFunction(){
  // Comentário em uma linha
}
```

> ### Comentário que inicia com ```/*``` e termina com ```*/```

O outro comentário pode ser utilizado para comentar várias linhas.
  ```
  function myFunction(){
    /* Esse comentário
    pode te mais de uma linha */
  }
  ```


> ## Importância da identação

> ### O que é identação? 

Identação é o padrão de escrita para um script. Normalmente adotando espaços e/ou Tabs para podermos identificar certos níveis hierárquicos de nosso código.

Em Javascript, um código pode ser escrito de qualquer maneira, pois a máquina não interpreta os espaços que damos em nosso código. Mas então por que precisamos da identação?

**Vamos ao exemplo: **

Peguemos o código anterior:

```
  function myFunction(){
    /* Esse comentário
    pode te mais de uma linha */
  }
```
Poderíamos re-escrever essa função da seguinte maneira:

``` 
function myFunction(){/*Esse comentário pode ter mais de uma linha*/}
```
Esse código é pequeno, mas já dá para ter uma noção de que um código mal identado, acaba sendo mais cansativo de se ler. Isso não é bom para nós programadores, pois dificulta nossa vida ao entender um código.

Sei que até esse código nós não vimos muita coisa de programação, mas vou deixar aqui um exemplo de um código bem identado e mal identado. Tente identificar qual o código com boa identação.

```
function primo(){
  var numero = 31;
  var divisores = 0;

  for(var count=1 ; count<=numero ; count++)
   if(numero % count == 0)
   	divisores++;
  
  if(divisores == 2)
  	Logger.log('É primo');
  else
  	Logger.log('Não é primo');
}

```

**OU**

```
function primo(){var numero = 31;var divisores = 0;
for(var count=1 ; count<=numero ; count++){if(numero % count == 0){divisores++;}}
if(divisores == 2){Logger.log('É primo');}else{Logger.log('Não é primo');}}
```

Um código sem indentação não é legal de se ler, mas tem uma coisa um pouco pior que isso, ou tão ruim quanto, que é a falta de um padrão na indentação do projeto.

**É MUITO importante padronizar o uso da indentação para não ter problemas futuros.**


> ## Variáveis


Uma variável é um container para um valor, como um número que podemos usar em uma operação de adição, ou uma sequência de texto que possamos usar como parte de uma oração.

Mas uma coisa especial a respeito das variáveis é que seu conteúdo pode mudar. Vamos ver um exemplo simples:

```
var n = 1
var n = 2 
```

O resultado para n será 2, pois o valor atribuido que antes era 1, agora é 2.

### Tipos de Variáveis

O Javascript possui sua tipagem dinâmica. Isso significa que não precisamos informar o tipo da variável ao declarar ela, o próprio JS consegue saber que tipo é o valor atribuído à variável e fazer o casting automático.


* Strings — Uma String nada mais é que texto puro.
* Numbers — São os números, seja eles integer, float, double etc.
* Booleans — São os operadores booleanos (true ou false)
* Arrays — É uma estrutura de dado para armazenar uma coleção de valores, sendo eles de qualquer tipo.
* Objects — Conjunto de atributos aninhados a uma variável denomina-se um objeto.
* Functions — Em JavaScript é possível declarar uma variável como uma função, podendo fazer operações e retornando o valor para a variável de declaração. Obs: muito utilizado no paradigma de programação funcional.

### Regras para nome de variáveis:

> * Não começar com números;

> * Não conter acentos;

> * Segunda palavra começar com letra maiúscula; 


``` 
function runScripts(){
  // Números
  var n_01 = 1      // inteiros - number
  var n_02 = 2      // float - number
  var n_03 = 1/2    // float - number

  var string = "teste 01"               // string
  var objetos = {"chave":"valor"}       // object / dicionário 
  var vector = ["teste 01","Teste 02"]  // Arrays

  // bools --> Booleano --- 0 ou 1
  var x = true
  var y = false

  var nulo = null
  var indefinido = undefined
}

```

```
function canIWork(teclado,mouse){
  return teclado && mouse
}
```  

exemplo de entradas:

```
var mouse = true
var teclado = false
```

Se chamarmos a função e passarmos ela como argumento para Logger.log para conseguir visualizar seu resultado:

```
function runScripts(){
  var mouse = true
  var teclado = false
  Logger.log(canIWork(teclado,mouse))
}
```

```
Response: false

```


#### Operador Lógico OR: || 

OR em inglês significa OU.

Exemplo: Posso *estudar por conta própria* ou *fazer uma faculdade* para aprender a programar

A função a seguir tem como argumentos estudarContaPropria e faculdade, e retornará um booleano indicando se eu conseguirei aprender ou não a programar.


```
function canILearnProgramming(estudarContaPropria,faculdade){
  return estudarContaPropria || faculdade
}
```

exemplo de entradas:
```
var estudarContaPropria = true
var faculdade = false
```

Se chamarmos a função e passarmos ela como argumento para Logger.log para conseguir visualizar seu resultado:

```
function runScripts(){
  var estudarContaPropria = true
  var faculdade = false
  Logger.log(canILearnProgramming(estudarContaPropria,faculdade))
}
```

```
Response: true

```


#### Operador lógico NOT: !

NOT em inglês significa não, é a negação.
Sempre que colocamos esse operador perto de alguma condição, ele altera o valor dela de verdadeiro para falso (true para false) ou o contrário, de falso para verdadeiro.

Exemplo:

```
function runScripts(){
  var verdadeiro = true
  var verdadeiro = !true
  Logger.log(verdadeiro)
}
```

```
Response: false
```

### Diferença entre “==” e “===”

Se usarmos o operador “==” em uma condição, saberemos que ela só será verdadeira se o valor da esquerda for o mesmo valor da direita, por exemplo:

```
function runScripts(){
  Logger.log(10 == '10')
}
```

Contudo, se precisarmos ter certeza de que tal variável é idêntico, ou seja, possui o mesmo valor e contém o mesmo tipo, utilizaremos o "==="


```
 function runScripts(){
  Logger.log(10 === '10')
}
```
